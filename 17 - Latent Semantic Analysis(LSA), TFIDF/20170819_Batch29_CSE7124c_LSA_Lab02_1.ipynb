# Computing tfidf

```python
vectorizer = TfidfVectorizer(min_df = 1,lowercase=True,ngram_range=(1,1) ,stop_words = 'english') 
dtm = vectorizer.fit_transform(example) 
Df=pd.DataFrame(dtm.toarray(),index=example,columns=vectorizer.get_feature_names())
Df
```

# #Computing SVD

```python

lsa = TruncatedSVD(6, algorithm = 'arpack') 
dtm_lsa = lsa.fit_transform(dtm) 
dtm_lsa = Normalizer(copy=False).fit_transform(dtm_lsa)
```

```python
dtm_lsa
```

```python
pd.DataFrame(lsa.components_,index = ["c1","c2","c3","c4","c5","c6"],columns = vectorizer.get_feature_names())
```

```python
pd.DataFrame(dtm_lsa, index = example, columns = ["c1","c2","c3","c4","c5","c6"])
```

```python
S=lsa.explained_variance_
S=numpy.sort(S,axis=0)
Sreverse=S[::-1]
Sreverse
```

```python
print(lsa.components_.shape)
print(dtm_lsa.shape)
print((lsa.explained_variance_))
print(Sreverse)
```

```python
a1=numpy.diag(Sreverse)
print(a1)
a2=numpy.zeros((6,6))
a2[0:4,0:4]=a1[0:4,0:4]
print(a2)
```

```python
a3=numpy.dot(dtm_lsa,a2)
a4=numpy.dot(a3,lsa.components_)
featurenames=vectorizer.get_feature_names()
import pandas as pd
df=pd.DataFrame(a4,columns=featurenames,index=[1,2,3,4,5,6,7])
```
